## 2 Arten zu plotten

```python
fig = plt.figure(figsize = (10,5))
ax = fig.add_subplot(111)
ax.plot([1, 2, 3, 4], [10, 20, 25, 30], color='lightblue', linewidth=3)
ax.scatter([0.3, 3.8, 1.2, 2.5], [11, 25, 9, 26], color='darkgreen', marker='^')
ax.set_xlim(0.5, 4.5)
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [10, 20, 25, 30], color='lightblue', linewidth=3)
plt.scatter([0.3, 3.8, 1.2, 2.5], [11, 25, 9, 26], color='darkgreen', marker='^')
plt.xlim(0.5, 4.5)
plt.show()
```

### Unterschiede

* Die zweite Variante sieht cleaner aus
* durch `plt.plot()` müssen wir die Achsen nicht explizit erzeugen
* Im Hintergrund wird bei `plt.xlim()` jedoch `ax.set_lim()` aufgerufen
* Die erste Variante ist trotzdem hilfreich z.B. wenn mehrere axes in einem Plot benutzt werden

**Anmerkung:**

* Achsen Limits lassen sich über `.set_xlim()` bzw. `y_lim` setzen
* Einzelne Punkte lassen sich über scatter einzeichnen 
* Punkte außerhalb der Achsen-Limits werden ignoriert 
* Setzt man das Limit nicht so werden die Achsen so angepasst, dass alle Daten im Plot sind

```python
plt.plot([1, 2, 3, 4], [10, 20, 25, 30], color='lightblue', linewidth=3)
plt.scatter([0.3, 3.8, 1.2, 2.5], [11, 25, 9, 26], color='darkgreen', marker='^')

plt.show()
```

```python
plt.plot([1, 2, 3, 4], [10, 20, 25, 30], color='lightblue', linewidth=3)
plt.scatter([0.3, 3.8, 1.2, 2.5], [11, 25, 9, 26], color='darkgreen', marker='^')
plt.xlim(0.5, 4.5)
plt.show()
```

### Bearbeiten von Plots

```python
x = np.linspace(0, 2, 100)

plt.plot(x, x, label='linear')  #Plotte die Gerade indem wir sowohl der x-Achse als auch der y-Achse die gleichen Daten übergeben, 
#Kreiere hierfür die nicht existente Achse
plt.plot(x, x**2, label='quadratic')  #plotte auf die selbe Achse
plt.plot(x, x**3, label='cubic') #usw.
plt.xlabel('x Achse') #Bennene die x-Achse
plt.ylabel('y Achse') #Bennene die y-Achse
plt.title("Beispiel Plot") #Gebe dem Plot einen Namen
plt.legend(loc='best') #Lasse die Legende ausgeben
plt.show() 
```

**Was ist ein Subplot?**

* Mit Subplot kann man Achsen in unserem Grid platzieren 
* Hat bis zu 3 Argumente als Input 1. die Anzahl der Reihen, 2. die Anzahl der Spalten und 3. die Anzahl der Plots

**Wie kann ich die Größe der Figure beeinlussen?**

* Das Argument figsize beim Erstellen des Plots übernimmt diese Aufgabe

```python
#Beachtet die Zahl der Größe
fig = plt.figure(figsize=(20,10)) 
fig, ax = plt.subplots()
```

### Axes

* Axes sind das, was man sich als Plot vorstellt 
* Eine Figur kann mehrere Axes als Objekt beinhalten, jedoch kann jede Axes nur in einem Plot vorhanden sein
* Axes beinhalten einzelne Axis (in 2-D 2 und in 3-D 3)
* Jede Axes lässt sich einen Namen über `plt.xlabel()` und `plt.ylabel()` zuweisen
* Außerdem 

### Axis 

*  bestimmt die Daten Limits
* bestimmt die einzelnen Ticks (Abschnitte) 

## Erstellen von Plots

### Basic Befehle:

Befehle     | Was passiert | 
:-|:-:|-: 
ax.bar()    | vertikale Rechtecke
ax.hbar()   | horizontale Rechtecke
ax.axhline()| horizontale Linie gegenüber der Achse
ax.vline()  | vertikale Linie gegenüber der Achse 
ax.fill()   | Ausfüllen der Polygone
ax.fill_between() | Ausfüllen der Fläche zwischen der y-Achse und 0
ax.stackplot() | Stack Plot


```python
# Initialize the plot
fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(131)
ax2 = fig.add_subplot(132)
ax3 = fig.add_subplot(133)


# Plot the data
ax1.bar([1,2,3],[3,4,5])
ax2.barh([0.5,1,2.5],[0,1,2])
ax2.axhline(0.45)
ax1.axvline(0.65)
ax3.scatter(x,x)
ax3.fill_between(x,x)

# Show the plot
plt.show()
```

### Visualisierungsbeispiel: Vektorfeld





```python
x = np.arange(-5,6,1)
y = np.arange(-5,6,1)

X, Y = np.meshgrid(x, y)

u, v = X/5, -Y/5

fig, ax = plt.subplots(figsize=(7,7))
ax.quiver(X,Y,u,v)

ax.xaxis.set_ticks([])
ax.yaxis.set_ticks([])
ax.axis([-6, 6, -6, 6])
ax.set_aspect('equal')

plt.show()
```

**Anmerkung:**

* Matplotlib nimmt am liebsten numpy arrays als Input (bzw. Numpy Array ähnliche Strukturen wie zum Beispiel pandas.Series)
* müssen dementsprechend oft Daten transformieren, um sie plotten zu können

### Weitere nützliche Features

**Achsen löschen**

```python

# Initialisierung des plots
fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(131)
ax2 = fig.add_subplot(132)
ax3 = fig.add_subplot(133)

# Plotten der Daten
ax1.bar([1,2,3],[3,4,5])
ax2.barh([1,2,3],[0,1,2])
ax2.axhline(4)
ax1.axvline(0.65)
ax3.scatter(np.linspace(0, 1, 5), np.linspace(0, 5, 5))

 #Löschen von `ax2`
fig.delaxes(ax2)

plt.show()
```

**Anpassen der einzelnen Plots**

* Bisher über die Größe der Figur gearbeitet
* Bevor `plt.show()` aufgerufen wird, nutze `plt.tight_layout()`, um die Plots schön anzupassen
* Auch über `subplots_adjust()` möglich, um manuell die Abstände zwischen den Plots einzurichten

```python
# Initialisierung des plots
fig = plt.figure()
ax1 = fig.add_subplot(131)
ax2 = fig.add_subplot(132)
ax3 = fig.add_subplot(133)

# Plotten der Daten
ax1.bar([1,2,3],[3,4,5])
ax2.barh([1,2,3],[0,1,2])
ax2.axhline(4)
ax1.axvline(0.65)
ax3.scatter(np.linspace(0, 1, 5), np.linspace(0, 5, 5))

 
plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(2,2)
fig.subplots_adjust(wspace=0.5,hspace=0.5) 
```

**Anmerkung:**

* über `wspace` setzen wir die Weite zwischen den Plots (zwischen den Spalten)
* über `hspace` setzen wir den Abstand zwischen den Zeilen
* hat außerdem die Inputs left, right, bottom und top für die jeweiligen Abstände

### Speichern und Schließen von Plots



** Speichern als Bild Datei**

* Nutze den Befehl `plt.savefig()`

```python
fig = plt.figure()
ax1 = fig.add_subplot(131)
ax2 = fig.add_subplot(132)
ax3 = fig.add_subplot(133)

# Plotten der Daten
ax1.bar([1,2,3],[3,4,5])
ax2.barh([1,2,3],[0,1,2])
ax2.axhline(4)
ax1.axvline(0.65)
ax3.scatter(np.linspace(0, 1, 5), np.linspace(0, 5, 5))

 
plt.tight_layout()
plt.savefig('Beispiel.png')
plt.close() # Schließt das Fenster mit dem Plot
```
